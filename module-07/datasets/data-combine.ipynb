```python
# Import
import pandas as pd

pd.set_option("display.max_columns", None)

spam_dataset1 = pd.read_csv("./email-spam-dataset.csv", encoding="latin1")
spam_dataset2 = pd.read_csv("./sms-spam-dataset.csv", encoding="latin1")
```

```python
def preprocess(email, sms):
    # Preprocess Email DataFrame
    email.rename(columns={"email": "message", "label": "spam"}, inplace=True)
    email['message_type'] = 'email'

    # Preprocess SMS DataFrame
    sms.drop(["Unnamed: 2", "Unnamed: 3", "Unnamed: 4"], axis=1, inplace=True)
    sms.rename(columns={"v1": "spam", "v2": "message"}, inplace=True)
    sms["spam"] = sms["spam"].map({"ham": 0, "spam": 1})
    sms['message_type'] = 'sms'

    # Combine the two DataFrames
    df = pd.concat([email, sms], axis=0, ignore_index=True)

    # Shuffle the combined DataFrame
    df = df.sample(frac=1, random_state=42).reset_index(drop=True)

    # Rearrange the columns
    df = df[["message", "message_type", "spam"]]

    return df
```

```python
df = preprocess(spam_dataset1, spam_dataset2)
df
```

```python
df.to_csv("spam-email+sms-dataset.csv", index=False)
```
