```python
# Importing
import pandas as pd
import altair as alt
import numpy as np

from IPython.display import Markdown
from IPython.display import display
from tabulate import tabulate
```

```python
# Load dataset
data = pd.read_csv('shoe-size-samples.csv')
data
```

```python
# Assuming you have a DataFrame 'data' containing your data
chart = alt.Chart(data).encode(
    x=alt.X('shoe_size',
            axis=alt.Axis(format='d', title='Shoe Size'),
            scale=alt.Scale(domain=[34, 52])),  # Set the domain for x-axis
    
    y=alt.Y('height', 
            axis=alt.Axis(format='d', title='Height'),
            scale=alt.Scale(domain=[140, 220])),  # Set the domain for y-axis
    
    color=alt.Color('sex:N')  # Color points based on the 'sex' variable
).mark_point()

chart
```
