```python
# Importing
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

pd.set_option("display.max_columns", None)


df = pd.read_csv(
    "https://raw.githubusercontent.com/byui-cse/cse450-course/master/data/housing.csv"
)

crime_rate = pd.read_csv("./crime_rate.csv")
```

```python
# Add crime rates from nearest city
df = pd.merge(df, crime_rate, on="zipcode", how="left")
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x="property_crime_rate_closest_city", y="price", color="green")
plt.title("Correlation between Property Crime Rates and Prices (DataFrame)")
plt.xlabel("Property Crime Rate")
plt.ylabel("Price")
plt.grid(True)
sns.despine()  # Removes the top and right spines for a cleaner look
plt.show()
```
