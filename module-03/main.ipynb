```python
# Import
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from pycaret.classification import *
from datetime import datetime

data = pd.read_csv(
    "https://raw.githubusercontent.com/byui-cse/cse450-course/master/data/housing.csv"
)
```

```python
data
```

```python
def transform_date(date_str):
    return pd.to_datetime(date_str, format="%Y%m%dT%H%M%S")


# Apply the transformation to all date columns
for column in data.columns:
    if "date" in column.lower():  # Assuming date columns have 'date' in their name
        data[column] = data[column].apply(transform_date)
```
